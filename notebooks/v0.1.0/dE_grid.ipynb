# Create a grid of dwarf galaxies as a function of mass, distance, age, and metallicity

```python
import numpy as np
import make_galaxies_grid
from astropy.convolution import convolve_fft
from inspect_galaxies import inspect, inspectall, inspectsum
```

```python
# Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sim = make_galaxies_grid.SimulateDwarfElliptical(
    distance=10.,
    logmass=6.,
    age=12.,
    feh=-1.75)

print("mass: ",np.log10(sim.mass))
print("distance: ",sim.distance)
print("re_kpc, arcsec: ",sim.re_kpc, sim.re_arcsec)
print("[Fe/H]: ",sim.feh)
print("age: ",sim.age/1.e9)
print("ellip: ",sim.ellipticity)
print("pa: ",sim.position_angle)
```
